# Estruturas de Repetição(For e While)

```python
"""
Estruturas de repetição são utilizadas para realizar uma mesma tarefa várias vezes,
evitando que tenhamos que escrever muito

Por exemplo, quero somar 2 números 100 vezes, eu não preciso escrever a mesma operação 100 vezes

While : Enquanto a condição for verdadeira, a operação será realizada
For: Itera por uma sequencia bem definida, por exemplo, no exemplo da soma usariamos um for que iteraria 100 vezes
"""
```

```python
valor = 1
while valor < 6:
  print(valor)
  valor += 1 # a cada loop ele printa e adiciona mais um ao valor


# sempre lembre-se de colocar alguma forma de sair do loop, senão cairá em um loop infinito
```

```python
valor = 1
while valor < 6:
  print(valor)
  valor += 1
  if valor == 4: # quando o valor se torna quatro ele para
    break
```

```python
"""
While True
um loop infinito que roda até encontrar um break
"""
soma = 0
while True : # um loop infinito, ele não para sobre nenhuma condição, apenas com o break
  digito = int(input("digite número a ser somado ou digite 1 para parar: "))
  if digito == 1 :
    break
  else:
    soma += digito



print(soma)
```

# Funções

```python
"""
Uma função é resumidamente um trecho de código que podemos reutilizar ao longo de
nosso código. Elas servem para que tenhamos que escrever menos e evitam o uso do
famoso Ctrl C Ctrl V.

O python já contem várias funções prontas como o sum(),len() e mean()
"""
```

```python
"""
Vou utilizar aquele mesmo problema das notas que eu fiz anteriormente, porém com uma leve modificação.
Dessa vez, quero converter uma lista de dez notas ao modelo americano
"""
notas = [100,60,50,30,90,80,50,70,80,100]

def nota_americana(nota):  # perceba que a função é criada antes e depois podemos chama-la globalmente no código
  if nota > 90:
   nota = "A"

  elif nota <= 90 and nota >=80:
   nota = "B"

  elif nota <=70:
   nota = "C"

  return nota
# nesse exemplo temos uma função que recebe um número e retorna um caractere, porém podemos ter outros modelos de função
for nota in notas:
  print(f"a nota foi {nota_americana(nota)}")
```

```python
"""
Função sem parametro, ela não recebe nenhum argumento e retorna alguma coisa
"""
def gol_da_alemanha():
  print("e é gol da alemanha")

gol_da_alemanha()
```

```python
"""
*arg: podemos adicionar um * antes de um parametro, caso não seja conhecido o número de argumentos que serão passados
**kwargs : podemons adicionar ** para aceitar um número variavel de argumentos chave pra função
"""
def somar_numeros(*numeros):
  return sum(numeros)

somar_numeros(20,10,30,40,100,100) # perceba que a função aceita um número variável de argumentos


def apresenta_membro(**kwargs):  # a função passa a aceitar um número variável de argumentos, no padrão de um dicionário (chave = valor)
  for chave,valor in kwargs.items():
    print(f"{chave}:{valor}")

apresenta_membro(nome = "Barata", idade = 26, cargo = "Egresso")
```

# Fundamentos de orientação a objetos

```python
"""
O que é a orientação a objetos:
Orientação a objetos (OO) é um paradigma de programação centrado no conceito de "objetos",
que são entidades que encapsulam dados e comportamentos. A ideia central
é representar conceitos do mundo real ou modelos como objetos que interagem entre si.

fazemos isso usando classes e objetos:
Classe : uma classe é basicamente um molde para a criação de um objeto. Ela define os atributos e
o que o objeto poderá fazer.

Objeto: um objeto é uma instância de uma classe.





Vantagens:
- Melhor organização do código
- Melhor controle sobre os usos das "Funções"

"""
```

```python
class Carro:
  def __init__(self,modelo,data_de_compra,): # isso aqui são os atributos da classe associado ao carro
    self.modelo = modelo
    self.data_de_compra = data_de_compra


chevrolet = Carro("Fiat Uno","02/10/2024")
chevrolet.data_de_compra

```

```python
"""
A parte mais interessante da orientação a objetos são os métodos. Um método é um trecho de código
que pode ser chamado por um objeto daquela respectiva classe. Os métodos funcionam como funções,
porém realizando alterações nos atributos do objeto.
"""

class Banco:
  def __init__(self,nome,saldo):
    self.nome = nome
    self.saldo = saldo
  def entrada_saldo(self,valor):
    self.saldo += valor
  def saida_saldo(self,valor):
    self.saldo -= valor

nubank = Banco("nubank",1000)
nubank.entrada_saldo(1000)
nubank.saldo

nubank.saida_saldo(500)
nubank.saldo
```

```python
"""
Método apply()

com esse método, podemos aplicar uma função nos valores de alguma coluna
"""
brasileirao3 = pd.read_csv("/content/campeonato-brasileiro-gols.csv")
brasileirao.dropna(how = "all")
brasileirao3.info()
brasileirao.head(25)


def tempo_de_jogo(tempo):
    if pd.isna(tempo) or not isinstance(tempo, str):
        return tempo
    tamanho = len(tempo)
    if tamanho <= 2:
        return int(tempo)
    else:
        lista_num = tempo.split("+")
        lista_int = [int(item) for item in lista_num]
        return sum(lista_int)


brasileirao3['minuto'] = brasileirao3['minuto'].apply(tempo_de_jogo) # nos aplicamos nossa função a cada um dos valores em cada linha dessa coluna
brasileirao3.head(30)
```

# Plotando dataframes usando matplotlib

## O que é o Matplotlib?


---
Matplotlib é uma biblioteca utilizada para expressar dados por meio de gráficos e tabelas. Seu grande objetivo é facilitar a visualização de métricas e informações presentes em uma base de dados por meio desses elementos mais visuais.

Há duas abordagens para se plotar gráfico utilizando matplotlib, uma utilizando funções para isso, e outra partindo com base na orientação a objetos. Nessa capacitação utilizaremos a forma orientada a objetos pois acredito que seja a que permite maior personalização e seja mais didatica(também é a minha favorita >:))

```python
!pip install matplotlib
import matplotlib.pyplot as plt
import pandas as pd
brasileirao = pd.read_csv("/content/Classificacao_2023.csv")
brasileirao

```

```python
"""
Fig(figure) - A figura total, onde podemos colocar vários gráficos
todos os métodos e atributos que mexem com a figura global são aplicados no fig

Ax(axes) - objeto para plotagem do gráfico(a parte de dentro onde realmente colocamos as informações)
todos os métodos e atributos que mexem apenas no gráfico especifico é feito no axes
"""
fig, ax = plt.subplots() # o .subplots() cria uma tela "vazia", para que possamos trabalhar nela

```

```python
fig, ax = plt.subplots()
palmeiras = brasileirao[brasileirao["Time"] == "Palmeiras"]
botafogo = brasileirao[brasileirao["Time"] == "Botafogo"]
ax.plot(palmeiras["Rodada"],palmeiras["Pontos"]) # respectivamente colocamos o x e o y do gráfico
ax.plot(botafogo["Rodada"],botafogo["Pontos"]) # podemos plotar varios graficos no mesmo ax
```

```python
"""
Para deixar mais claro, vamos adicionar um titulo ao gráfico e nomeações para os eixos
"""

fig, ax = plt.subplots()
ax.set_title("Palmeiras x Botafogo ao longo das rodadas",fontsize = 14) # adicionando um titulo
ax.set_xlabel("Rodada") # label do eixo x
ax.set_ylabel("Pontos") # label do eixo y
ax.plot(palmeiras["Rodada"],palmeiras["Pontos"],color = "green") # respectivamente colocamos o x e o y do gráfico
ax.plot(botafogo["Rodada"],botafogo["Pontos"], color = "black")
```

```python
"""
Podemos melhorar ainda mais com uma legenda para o nosso gráfico
"""
fig, ax = plt.subplots()
ax.set_title("Palmeiras x Botafogo ao longo das rodadas",fontsize = 14) # adicionando um titulo
ax.set_xlabel("Rodada") # label do eixo x
ax.set_ylabel("Pontos") # label do eixo y
ax.plot(palmeiras["Rodada"],palmeiras["Pontos"],color = "green",label = "Palmeiras") # adicionamos os labels como parametro
ax.plot(botafogo["Rodada"],botafogo["Pontos"], color = "black", label = "Botafogo")
plt.legend(loc = "lower right") # .legend() para mostrar a legenda / loc para definir o local da legenda
# ou bbox_to_anchor = (1,0.3) para posições exatas
plt.show()
```
